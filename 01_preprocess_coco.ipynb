```python
def load_image_and_mask(path):
    """
    Load a multi-band GeoTIFF exported from GEE and build a valid-pixel mask.
    
    Returns
    -------
    img : np.ndarray
        Array of shape (bands, H, W)
    valid_mask : np.ndarray
        Boolean array of shape (H, W), True = valid pixel
    meta : dict
        Raster metadata (optional)
    """
    with rasterio.open(path) as src:
        img = src.read().astype(np.float32)
        meta = src.meta.copy()

    # GEE float exports â†’ masked pixels = NaN
    valid_mask = ~np.any(np.isnan(img), axis=0)

    return img, valid_mask, meta
```

```python
# Apply mask to all images

data = {}

for fname in tif_files:
    path = os.path.join(images_path, fname)
    img, mask, meta = load_image_and_mask(path)

    data[fname] = {
        "image": img,
        "valid_mask": mask,
        "meta": meta,
        "missing_ratio": 1 - mask.mean()
    }

    print(f"{fname}: missing {1 - mask.mean():.2%}")
```

## Training set

```python
def compute_normalization_stats(train_images):
    """
    train_images: list of (img, valid_mask) tuples
    img shape: (bands, H, W)
    """
    band_values = [[] for _ in range(train_images[0][0].shape[0])]

    for img, mask in train_images:
        for b in range(img.shape[0]):
            band_values[b].append(img[b][mask])

    means = np.array([np.mean(np.concatenate(v)) for v in band_values])
    stds  = np.array([np.std(np.concatenate(v))  for v in band_values])

    return means, stds
```

### Split

```python
#Patch extraction from raw images + mask
print("Change PATCH size here if needed")
PATCH = 32

```

```python

def extract_patches(img, mask, patch_size=PATCH, stride=PATCH//2, valid_thresh=0.9, source=None):
    B, H, W = img.shape
    patches = []

    for y in range(0, H - patch_size + 1, stride):
        for x in range(0, W - patch_size + 1, stride):
            patch = img[:, y:y+patch_size, x:x+patch_size]
            #patch_mask = mask[y:y+patch_size, x:x+patch_size]
            if np.isnan(patch).any():
                continue


            #if patch_mask.mean() >= valid_thresh:
            patches.append({
                "data": patch,
                "y": y,
                "x": x,
                "source": source                                })

    return patches
```

```python
# Apply to all training images
train_files = [
    ("L5_Glacier_train_2010_30m.tif", "2010"),
    ("L8_Glacier_train_2016_30m.tif", "2016"),
]

train_patches = []
train_imgs = []

for fname, year in train_files:
    img, mask, _ = load_image_and_mask(os.path.join(images_path, fname))
    
    # Harmonize bands between Landsat 5 and Landsat 8
    # for both, we keep only R, G, B and SWIR1 
    bands_to_keep = [0, 1, 2, 4]  # Example: keep bands 0, 1, 2 and 4 (R, G, B, SWIR1)
    if fname.startswith("L5"):
        data[fname]["image"] = data[fname]["image"][bands_to_keep, :, :]
        # inverse the band indices to have same order as Landsat 8 
        data[fname]["image"] = data[fname]["image"][[2, 0, 1, 3], :, :]
    if fname.startswith("L8"):
        data[fname]["image"] = data[fname]["image"][bands_to_keep, :, :]

    img = data[fname]["image"]

    train_imgs.append((img, mask))

    patches = extract_patches(
        img,
        mask,
        patch_size=PATCH,
        stride=PATCH//2,
        source=year   # THIS must never be None
    )

    train_patches.extend(patches)

# compute statistics after harmonization
means, stds = compute_normalization_stats(train_imgs)
print("Band means after harmonization:", means)
print("Band stds after harmonization:", stds)

print("Total TRAIN patches:", len(train_patches))
```

### Label (rasterize + slice)

```python
def rasterize_glacier_binary(shp_path, reference_raster_path):
    """
    Returns a binary glacier mask:
    1 = glacier
    0 = non-glacier
    """
    gdf = gpd.read_file(shp_path)

    with rasterio.open(reference_raster_path) as src:
        transform = src.transform
        height = src.height
        width = src.width
        crs = src.crs

    # Ensure CRS match
    if gdf.crs != crs:
        gdf = gdf.to_crs(crs)

    shapes = [(geom, 1) for geom in gdf.geometry]

    glacier_mask = rasterize(
        shapes=shapes,
        out_shape=(height, width),
        transform=transform,
        fill=0,
        dtype=np.uint8
    )

    return glacier_mask

```

```python
glacier_mask_2010 = rasterize_glacier_binary(
    glacier_2010_path,
    os.path.join(images_path, "L5_Glacier_train_2010_30m.tif")
)

glacier_mask_2016 = rasterize_glacier_binary(
    glacier_2016_path,
    os.path.join(images_path, "L8_Glacier_train_2016_30m.tif")
)

```

```python
train_patches_clean = []
Y_train = []


for p in train_patches:
    if p["source"] == "2010":
        glacier_mask_ref = glacier_mask_2010
    elif p["source"] == "2016":
        glacier_mask_ref = glacier_mask_2016

   
    patch_mask = glacier_mask_ref[
        p["y"]:p["y"]+PATCH,
        p["x"]:p["x"]+PATCH
    ]

    # (optionnel) filtrer les patches 100% fond
    # if patch_mask.sum() == 0:
    #     continue

    train_patches_clean.append(p)
    Y_train.append(patch_mask)

# remplacer les anciennes variables
train_patches = train_patches_clean
Y_train = np.stack(Y_train).astype(np.uint8)

X_train = np.stack([p["data"] for p in train_patches])


```

## Visual check

```python
import matplotlib.pyplot as plt
import numpy as np

def show_XY(X, Y, idx, bands=(0,1,2)):
    """
    X: (N, B, H, W)
    Y: (N, H, W)
    idx: index du patch Ã  afficher
    bands: bandes Ã  utiliser pour le RGB
    """
    x = X[idx]
    y = Y[idx]

    # Construire RGB
    rgb = x[list(bands)]
    rgb = (rgb - rgb.min()) / (rgb.max() - rgb.min() + 1e-6)
    rgb = rgb.transpose(1, 2, 0)

    fig, axes = plt.subplots(1, 3, figsize=(12, 4))

    axes[0].imshow(rgb)
    axes[0].set_title("X (image patch)")
    axes[0].axis("off")

    axes[1].imshow(y, cmap="gray")
    axes[1].set_title("Y (glacier mask)")
    axes[1].axis("off")

    axes[2].imshow(rgb)
    axes[2].imshow(y, cmap="Reds", alpha=0.45)
    axes[2].set_title("Overlay")
    axes[2].axis("off")

    plt.tight_layout()
    plt.show()
```
